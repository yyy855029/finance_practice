```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter,FormatStrFormatter
%matplotlib inline

#設定字體和解決圖像中文顯示問題
plt.rcParams['font.sans-serif']=['Microsoft JhengHei']
plt.rcParams['axes.unicode_minus']=False

#利用numpy和scipy
def future_bs(S,K,r,T,sigma):
    d1=(np.log(S/K)+(0.5*sigma**2)*T)/(sigma*np.sqrt(T))
    d2=d1-sigma*np.sqrt(T)
    C=np.exp(-r*T)*(S*norm.cdf(d1)-K*norm.cdf(d2))
    return C

st=np.linspace(11200,12200,1001)
k=11600
c=np.maximum(st-k,0)-133
ct=np.array([future_bs(S,11600,0.008731,7/365,0.1149)-133 for S in st])
profit=1490*50*(ct-1)

fig,axes=plt.subplots(2,figsize=(10,10))
axes[0].plot(st,c,linestyle='-',label='內含價值')
axes[0].plot(st,ct,linestyle='-',label='內含價值+時間價值')
axes[0].spines['bottom'].set_position(('data',0))
axes[0].spines['top'].set_visible(False)
axes[0].spines['right'].set_visible(False)
axes[0].set_title('買進買權策略',fontsize=18)
axes[0].tick_params(axis='y',labelsize=8)
axes[0].tick_params(axis='x',labelsize=8)
axes[0].set_ylim(-200,400)
axes[0].set_xlim(11200,12200)
axes[0].set_xticks(np.arange(11300,12200,100))
axes[0].grid()
axes[0].legend(bbox_to_anchor=(1.01,1),loc='upper left',fontsize=12)


axes[1].plot(st,profit,label='損益')
axes[1].spines['bottom'].set_position(('data',0))
axes[1].spines['top'].set_visible(False)
axes[1].spines['right'].set_visible(False)
axes[1].set_title('預期下週損益',fontsize=18)
axes[1].yaxis.set_major_formatter(FormatStrFormatter('%.0f'))
axes[1].tick_params(axis='y',labelsize=8)
axes[1].tick_params(axis='x',labelsize=8)
axes[1].set_xlim(11200,12200)
axes[1].set_xticks(np.arange(11300,12200,100))
axes[1].grid()
axes[1].legend(bbox_to_anchor=(1.01,1),loc='upper left',fontsize=12)

#negative:True,positive:False
negative=profit<0
#profit負值個數
print('負值共有{}個，從編號0到{}'.format(negative.sum(),(negative.sum()-1)),'\n')

#st
print('最後一個出現負值的股價是編號{}，股價:{}'.format((negative.sum()-1),st[negative.sum()-1]))
print('第一個出現正值的股價是編號{}，股價:{}'.format((negative.sum()),st[negative.sum()]),'\n')

#profit
print('最後一個出現負值的損益是編號{}，損益:{}'.format((negative.sum()-1),profit[negative.sum()-1]))
print('第一個出現正值的損益是編號{}，損益:{}'.format((negative.sum()),profit[negative.sum()]),'\n')
```

```python
from io import StringIO
import requests
from datetime import datetime,timedelta
import numpy as np
import pandas as pd
from pandas.plotting import register_matplotlib_converters
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter
import matplotlib.dates as mdates
%matplotlib inline

register_matplotlib_converters()
#設定字體和解決圖像中文顯示問題
plt.rcParams['font.sans-serif']=['Microsoft JhengHei']
plt.rcParams['axes.unicode_minus']=False

#定義爬取FRED價格資料函數
def fred_price_crawler(ids,startdate,enddate):
    startdate=startdate.strftime('%Y-%m-%d')
    enddate=enddate.strftime('%Y-%m-%d')
    url='https://fred.stlouisfed.org/graph/fredgraph.csv?id={}&cosd={}&coed={}'.format(ids,startdate,enddate)
    response=requests.get(url)
    df=pd.read_csv(StringIO(response.text))
    df[ids]=pd.to_numeric(df[ids],errors='coerce')
    df['DATE']=pd.to_datetime(df['DATE'])
    
    return df

enddate=datetime.now().date()
startdate=(enddate-timedelta(days=365*5))

rate_30=fred_price_crawler('DGS30',startdate,enddate)
rate_2=fred_price_crawler('DGS2',startdate,enddate)
rate=pd.merge(rate_30,rate_2,on='DATE')
rate['DIFF']=rate['DGS30']-rate['DGS2']
rate.set_index('DATE',inplace=True)


fig,axes=plt.subplots(2,figsize=(12,10))
axes[0].spines['top'].set_visible(False)
axes[0].spines['right'].set_visible(False)
axes[0].plot(rate.index,rate['DGS2'],label='2年期美國公債殖利率(%)')
axes[0].plot(rate.index,rate['DGS30'],label='30年期美國公債殖利率(%)')
axes[0].axvline('2018-01-31',linestyle='--',color='dimgray')
axes[0].axvline('2018-07-31',linestyle='--',color='dimgray')
axes[0].fill_between(['2018-01-31','2018-07-31'],0,3.5,color='gray',alpha=0.1)
axes[0].set_title('美國公債殖利率(%)走勢圖',fontsize=15)
axes[0].legend(bbox_to_anchor=(1.01,1),loc='upper left',fontsize=11)
axes[0].xaxis.set_major_locator(mdates.MonthLocator(interval=6))
axes[0].tick_params(axis='x',labelsize=8)
axes[0].set_ylim(0,3.5)
axes[0].grid()

axes[1].spines['top'].set_visible(False)
axes[1].spines['right'].set_visible(False)
axes[1].plot(rate.index,rate['DIFF'],color='g')
axes[1].axhline(0.1023,linestyle='--',color='r')
axes[1].axvline('2018-01-31',linestyle='--',color='dimgray')
axes[1].axvline('2018-07-31',linestyle='--',color='dimgray')
axes[1].fill_between(['2018-01-31','2018-07-31'],0,3.5,color='gray',alpha=0.1)
axes[1].set_title('美國公債殖利率利差(%)走勢圖',fontsize=15)
axes[1].set_ylim(0,3.5)
axes[1].set_yticks([0.1023,0.5,1,1.5,2,2.5,3,3.5])
axes[1].yaxis.set_major_formatter(FormatStrFormatter('%.4f'))
axes[1].xaxis.set_major_locator(mdates.MonthLocator(interval=6))
axes[1].tick_params(axis='x',labelsize=8)
axes[1].grid()

plt.tight_layout()
```
